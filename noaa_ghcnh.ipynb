```python
import pandas as pd
import os

# Define the bounding box
lon_min = -98.46 - 1
lat_min = 30.18 - 1
lon_max = -98.46 + 1
lat_max = 30.18 + 1

# Directory containing the .psv files
directory = "/content"

# Initialize a list to store matching file names
matching_files = []

# Loop through each file in the directory
for file_name in os.listdir(directory):
    if file_name.endswith(".psv"):
        file_path = os.path.join(directory, file_name)

        # Read the file
        try:
            data = pd.read_csv(file_path, sep='|')

            # Check if any latitude and longitude fall within the bounding box
            if ((data['Latitude'] >= lat_min) &
                (data['Latitude'] <= lat_max) &
                (data['Longitude'] >= lon_min) &
                (data['Longitude'] <= lon_max)).any():
                matching_files.append(file_name)
        except Exception as e:
            print(f"Error processing {file_name}: {e}")

# Output the matching files
print("Files with lat/lon within the bounding box:")
for file in matching_files:
    print(file)
```

```python
import pandas as pd
import numpy as np
# Function to check if the column contains only NaN values
def is_column_all_nan(df, column_name):
    # Replace unexpected placeholders with proper NaN
    df[column_name] = df[column_name].replace(['', 'NaN', 'nan', 'N/A', 'null', None], np.nan)
    # Ensure the column is numeric or coerced to NaN where necessary
    df[column_name] = pd.to_numeric(df[column_name], errors='coerce')
    return df[column_name].isna().all()

# Process files
valid_files = []

for file in matching_files:
    try:
        # Read the file
        df = pd.read_csv(file, sep='|')
        print(f"Processing file: {file}")

        # Inspect unique values for debugging
        print(f"Unique values in 'precipitation': {df['precipitation'].unique()}")

        # Check if 'precipitation' column has only NaN
        if not is_column_all_nan(df, 'precipitation'):
            valid_files.append(file)  # Add to valid files
    except Exception as e:
        print(f"Error processing {file}: {e}")

# Output valid files
print("Files with valid precipitation data:")
for file in valid_files:
    print(file)
```
